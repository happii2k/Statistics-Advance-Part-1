# **Statistics Advance Part 1**
##**1. What is a random variable in probability theory?**

A random variable is a variable that takes numerical values based on the outcomes of a random experiment.

##**2. What are the types of random variables?**

Discrete random variable: Takes specific, countable values (e.g., 0, 1, 2).

Continuous random variable: Takes any value in a range (e.g., height, weight).

##**3. What is the difference between discrete and continuous distributions?**

Discrete distributions deal with countable outcomes.

Continuous distributions deal with measurable outcomes over intervals.

##**4. What are probability distribution functions (PDF)?**
A Probability Distribution Function (PDF) describes the likelihood of a random variable taking certain values.

* For discrete: It's called a Probability Mass Function (PMF).

* For continuous: PDF gives the density, not direct probabilities.

##**5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

CDF gives the probability that a variable is less than or equal to a value.

PDF (for continuous) shows the relative likelihood of different outcomes.

##**6. What is a discrete uniform distribution?**
A discrete uniform distribution assigns equal probability to each of a finite number of outcomes (e.g., rolling a fair die).

##**7. What are the key properties of a Bernoulli distribution?**

Only two outcomes: success (1) or failure (0).

Defined by a single parameter p (probability of success).

##**8. What is the binomial distribution, and how is it used in probability?**
Describes the number of successes in n independent Bernoulli trials with probability p of success in each.

##**9. What is the Poisson distribution and where is it applied?**
Models the number of times an event occurs in a fixed interval of time/space, given the average rate (λ).
Used for rare events (e.g., phone calls per minute).

##**10. What is a continuous uniform distribution?**
Every value within an interval [a, b] is equally likely. The PDF is constant over this interval.

##**11. What are the characteristics of a normal distribution?**

Bell-shaped and symmetric around the mean (μ).

Defined by mean (μ) and standard deviation (σ).

Most values fall within ±3σ of the mean.

##**12. What is the standard normal distribution, and why is it important?**
A normal distribution with mean 0 and standard deviation 1.
It's the basis for Z-scores and statistical inference.

##**13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
CLT states that the sampling distribution of the sample mean approximates a normal distribution as the sample size grows, regardless of the population's distribution.

##**14. How does the Central Limit Theorem relate to the normal distribution?**
It explains why normal distributions arise frequently, allowing for normal approximation in sampling and inference.

##**15. What is the application of Z statistics in hypothesis testing?**
Z-statistics help determine how far a sample statistic is from the hypothesized population parameter under the null hypothesis, in terms of standard deviations.

##**17. What are point estimates and interval estimates in statistics?**

Point estimate: A single value estimate (e.g., sample mean).

Interval estimate: A range of values likely to contain the population parameter (e.g., confidence interval).

##**18. What is the significance of confidence intervals in statistical analysis?**
A confidence interval gives a range within which the true population parameter is expected to lie with a certain level of confidence (e.g., 95%).

##**19. What is the relationship between a Z-score and a confidence interval?**
Z-scores determine the bounds of the confidence interval.
For example, a 95% CI uses Z ≈ ±1.96.

##**20. How are Z-scores used to compare different distributions?**
By standardizing values, Z-scores allow comparison across different distributions with different means and standard deviations.

##**21. What are the assumptions for applying the Central Limit Theorem?**

Sample size is large (typically n ≥ 30).

Samples are independent.

Population has a finite mean and variance.

##**23. How does a probability distribution relate to the expected outcome of a random variable?**
The distribution determines how likely each outcome is, and the expected value is the weighted average of all possible outcomes.
